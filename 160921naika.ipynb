```python
from urllib.request import urlopen
import re

def Get_Generalists (url):
    kengoto_list = []
    f = urlopen(url).read().decode('utf-8')
    p1 = re.compile (r'<td width="20%">(?P<drs>[\w\W]*?)</td>')
    for m in re.finditer(p1, f):
        kengoto_list.append(m.group('drs'))
    p2 = re.compile (r'<td>(?P<drs>[\w\W]*?)</td>')
    for m in re.finditer(p2, f):
        kengoto_list.append(m.group('drs'))
    return(kengoto_list)


```

```python

test_list = []
u =  'http://www.naika.or.jp/nintei/seido/meibo/s08/'

ret = Get_Generalists(u )

for m in ret:
    print(m)


```
