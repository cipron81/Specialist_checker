```python
from urllib.request import urlopen
import re

f = urlopen('http://www.naika.or.jp/nintei/seido/meibo/s01/')
```

```python
k = f.read().decode('utf-8')
#n<!-- コンテンツ開始 \n―――――――――――――――――――――――――――――――――――――――――――――――――――――――――-->\n
#から
#<!-- .entry-content -->
#までがデータ

print(k)
```

```python

p = re.compile (r'<div id="my_form_contents">[\w\W]*<div class="childPageBox">')


con= p.search(k)
con.group()
```
