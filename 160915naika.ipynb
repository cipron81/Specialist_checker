```python
from urllib.request import urlopen
import re

f = urlopen('http://www.naika.or.jp/nintei/seido/meibo/s01/')
```

```python
k = f.read().decode('utf-8')
#n<!-- コンテンツ開始 \n―――――――――――――――――――――――――――――――――――――――――――――――――――――――――-->\n
#から
#<!-- .entry-content -->
#までがデータ

print(k)
```

```python

p = re.compile (r'<div id="my_form_contents">[\w\W]*<div class="childPageBox">')


con= p.search(k)
con.group()
```

```python
#<td>名前</td>ってなってそう
p1 = re.compile (r'<td width="20%">(?P<drs>[\w\W]*?)</td>')

for m in re.finditer(p1, k):
    print(m.group('drs').encode('utf8'))

con1= p1.findall(k)
print(len(con1))
print(con1)

p2 = re.compile (r'<td>[\w\W]*?</td>')
con2= p2.findall(k)
print(len(con2))
print(con2)
```

```python
#同姓同名検索がいたら困るから_都道府県をつけようかな。

dr_list = []

p1 = re.compile (r'<td width="20%">(?P<drs>[\w\W]*?)</td>')
for m in re.finditer(p1, k):
    dr_list.append(m.group('drs'))

p2 = re.compile (r'<td>(?P<drs>[\w\W]*?)</td>')
for m in re.finditer(p2, k):
    dr_list.append(m.group('drs'))

    
for m in dr_list:
    print(m)

    
```
